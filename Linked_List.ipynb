```
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.previous = None
  
  def __repr__(self):
    return self.data

class LinkedList:
  def __init__(self, nodes):
    self.head = None
    if nodes is not None:
      node = Node(data=nodes.pop(0))
      self.head = node
      for elem in nodes:
        node.next = Node(data=elem)
        node = node.next
    else:
      print("Nodes can not be empty list")
  
  def __repr__(self):
    node = self.head
    nodes = []
    while node is not None:
      nodes.append(node.data)
      node = node.next
    nodes.append('None')
    return '-->'.join(nodes)

  def __iter__(self):
    node = self.head
    while node is not None:
      yield node
      node = node.next
  
  def length(self):

    if not self.head:
      raise Exception("List is empty!")

    counter = 0
    node = self.head
    while node is not None:
      counter += 1
      node = node.next
    return counter
  
  def get(self, index):

    if not self.head:
      raise Exception("List is Empty")
    
    if index >= self.length():
      raise Exception("Index is out of range!")
    
    pos = 0
    for node in self:
      if pos == index:
        return node.data
      pos += 1
    

  
  def add_first(self, node):
    node.next = self.head
    self.head = node
  
  def add_last(self, node):
    if not self.head:
      self.head = node
      return
    
    for cur_node in self:
      pass
    cur_node.next = node

  def insert_before(self, target_node_data, new_node):

    if target_node_data is None:
      raise Exception("Please enter target_node_data!")
    
    if new_node is None:
      raise Exception("Pleaes enter new node!")

    if not self.head:
      raise Exception("List is empty!")

    if self.head.data == target_node_data:
      return self.add_first(new_node)
    
    prev_node = self.head
    for node in self:
      if node.data == target_node_data:
        prev_node.next = new_node
        new_node.next = node
        return
      prev_node = node
  
  def insert_after(self, target_node_data, new_node):

    if target_node_data is None:
      raise Exception("Please enter target node data")
    
    if new_node is None:
      raise Exception("Pleaes enter new node")
    
    for node in self:
      if node.data == target_node_data:
        new_node.next = node.next
        node.next = new_node
        return
    
    raise Exception("The node was not found in the list")
  
  def remove_node(self, target_node_data):

    if not self.head:
      raise Exception("The list is empty")
    
    if self.head.data == target_node_data:
      self.head.next = self.head
    
    prev_node = self.head
    for node in self:
      if node.data == target_node_data:
        prev_node.next = node.next
        return
      prev_node = node
    
    raise Exception("The node does not exist!")

  def reverse(self):

    if not self.head:
      raise Exception("List is empty!")
    
    if self.length() == 1:
      return self.head
    
    prev_node = None
    cur_node = self.head
    while cur_node is not None:
      next = cur_node.next
      cur_node.next = prev_node
      prev_node = cur_node
      cur_node = next
    self.head = prev_node

class DoubleLinkedList:

  def __init__(self,nodes):
    self.head = None
    if nodes is not None:
      node = Node(data=nodes.pop(0))
      node.previous = None
      self.head = node
      for elem in nodes:
        new_node = Node(data=elem)
        node.next = new_node
        node.previous = node
        node = node.next
    else:
      print("Nodes is empty!")
    
  def __repr__(self):

    node = self.head
    nodes = []
    while node is not None:
      nodes.append(node.data)
      node = node.next
    nodes.append('None')
    return '<-->'.join(nodes)

  def __iter__(self):
    node = self.head
    while node is not None:
      yield node
      node = node.next
  
  def length(self):

    if not self.head:
      raise Exception("The list is empty!")

    counter = 0
    node = self.head
    for node in self:
      counter += 1
    return counter

  def get(self, index):

    if not self.head:
      raise Exception("List is empty!")
    
    if index >= self.length():
      raise Exception("Index is out of range!")

    node_index = 0
    for node in self:
      if node_index == index:
        return node.data
      node_index += 1

  def insert_begining(self, node):

    if not self.head:
      self.head = node
    
    if node is None:
      raise Exception("Please provide Node!")
    
    node.next = self.head
    self.head.previous = node
    self.head = node

  def insert_last(self,node):

    if not self.head:
      raise Exception("List is empty")
    
    if node is None:
      raise Exception("Please provide Node!")
    
    for cur_node in self:
      pass
    cur_node.next = node
    node.previous = cur_node

  



    

```
